```python
import sys
import os
sys.path.insert(0,'..')

import numpy as np
import matplotlib.pyplot as plt

import scipy.io
%matplotlib notebook
```

```python
dc1 = scipy.io.loadmat('../LQ_game/data/CoPG_lr0_1.mat')
dc2 = scipy.io.loadmat('../LQ_game/data/CoPG_lr0_01.mat')
dc3 = scipy.io.loadmat('../LQ_game/data/CoPG_lr0_005.mat')
dc4 = scipy.io.loadmat('../LQ_game/data/CoPG_lr0_001.mat')
dc5 = scipy.io.loadmat('../LQ_game/data/CoPG_lr0_0001.mat')

dg1 = scipy.io.loadmat('../LQ_game/data/GDA_lr0_01.mat')
dg2 = scipy.io.loadmat('../LQ_game/data/GDA_lr0_005.mat')
dg3 = scipy.io.loadmat('../LQ_game/data/GDA_lr0_001.mat')
dg4 = scipy.io.loadmat('../LQ_game/data/GDA_lr0_0001.mat')
```

```python
ax = plt.subplot()
ax.clear()
plt.xlabel('$Iterations$')
plt.ylabel('$\eta ( \\theta\'^2, \\theta\'^1) - \eta ( \\theta^2, \\theta^1) $')
ax.plot(dc2['a1'][0], label='CoPG: lr1e-2')
ax.plot(dg1['a1'][0], label='GDA: lr1e-2')
ax.plot(dc1['a1'][0], label='CoPG: lr1e-1')
ax.set_xlim([0, 80])
ax.set_ylim([-3, 8])
plt.legend(loc='upper right')
plt.show()
```
